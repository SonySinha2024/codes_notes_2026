```python
# Grouping by 'Depot_Country'
gcdf = df.groupby('Depot_Country')['Original_Total'].mean().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(x='Depot_Country', y='Original_Total', data=gcdf)
plt.title('Average Original_Total by Depot Country')
plt.xlabel('Depot Country')
plt.ylabel('Average Original_Total')
plt.show()
```
